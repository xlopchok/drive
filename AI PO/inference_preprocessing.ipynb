```python
import os
import torch
import numpy as np
from tqdm import tqdm  # For progress bar
from multiprocessing import Pool, cpu_count

def save_points_as_stl(points, output_path):
    """
    Save points as an STL file by creating degenerate triangles.
    Each point becomes a triangle with zero area.
    """
    num_points = points.shape[0]
    with open(output_path, 'w') as f:
        f.write('solid point_cloud\n')
        for i in range(num_points):
            vertex = points[i]
            # Create a degenerate triangle where all vertices are the same point
            f.write('  facet normal 0.0 0.0 0.0\n')
            f.write('    outer loop\n')
            f.write(f'      vertex {vertex[0]} {vertex[1]} {vertex[2]}\n')
            f.write(f'      vertex {vertex[0]} {vertex[1]} {vertex[2]}\n')
            f.write(f'      vertex {vertex[0]} {vertex[1]} {vertex[2]}\n')
            f.write('    endloop\n')
            f.write('  endfacet\n')
        f.write('endsolid point_cloud\n')

def process_pt_file(args):
    """
    Load a tensor from a .pt file and save it as an STL file with degenerate triangles.
    """
    file_path, output_dir = args
    try:
        # Load the tensor
        tensor = torch.load(file_path)

        # Ensure tensor is on CPU and convert to NumPy array
        tensor = tensor.cpu().numpy()

        # Check the shape of the tensor
        if tensor.ndim != 2 or tensor.shape[1] != 3:
            print(f"Skipping {file_path}: Tensor shape {tensor.shape} is not [N, 3]")
            return

        # Generate output file path
        filename = os.path.basename(file_path)
        output_filename = os.path.splitext(filename)[0] + '.stl'
        output_path = os.path.join(output_dir, output_filename)

        # Save points as STL
        save_points_as_stl(tensor, output_path)
        return f"Saved STL file to {output_path}"
    except Exception as e:
        return f"Error processing {file_path}: {e}"

def main(input_dir, output_dir):
    """
    Iterate over all .pt files in input_dir and save them as STL files in output_dir using multiprocessing.
    """
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Find all .pt files in the input directory
    pt_files = []
    for root, dirs, files in os.walk(input_dir):
        for file in files:
            if file.endswith('.pt'):
                pt_files.append(os.path.join(root, file))

    if not pt_files:
        print("No .pt files found in the specified directory.")
        return

    print(f"Found {len(pt_files)} .pt files.")

    # Prepare arguments for multiprocessing
    args_list = [(pt_file, output_dir) for pt_file in pt_files]

    # Use multiprocessing Pool to process files in parallel
    num_processes = min(cpu_count(), len(pt_files))
    print(f"Using {num_processes} processes for multiprocessing.")

    with Pool(processes=num_processes) as pool:
        # Use tqdm to display progress bar
        results = list(tqdm(pool.imap_unordered(process_pt_file, args_list), total=len(args_list), desc="Processing .pt files"))

    # Print results
    for res in results:
        if res:
            print(res)
```

```python
import os
import numpy as np
from stl import mesh
from multiprocessing import Pool, cpu_count

def txt_to_stl(input_txt, output_stl):
    try:
        # Load points from TXT file
        points = np.loadtxt(input_txt, delimiter=",")
        if points.shape[1] != 3:
            raise ValueError(f"File {input_txt} must contain exactly 3 values per line (x, y, z).")

        # Create degenerate triangles for STL (each point forms a degenerate triangle)
        num_points = points.shape[0]
        faces = np.zeros((num_points, 3, 3))
        for i in range(num_points):
            faces[i] = [points[i], points[i], points[i]]

        # Create an STL mesh
        degenerate_mesh = mesh.Mesh(np.zeros(num_points, dtype=mesh.Mesh.dtype))
        for i, f in enumerate(faces):
            degenerate_mesh.vectors[i] = f

        # Save the mesh to an STL file
        degenerate_mesh.save(output_stl)
        print(f"Saved degenerate STL to: {output_stl}")
    except Exception as e:
        print(f"Error processing {input_txt}: {e}")

def process_file(file_info):
    """Wrapper function for processing a single file."""
    input_txt, output_stl = file_info
    txt_to_stl(input_txt, output_stl)

def process_txt_files(input_dir, output_dir):
    # Ensure the output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Prepare file pairs for multiprocessing
    file_pairs = []
    for filename in os.listdir(input_dir):
        if filename.endswith(".txt"):
            input_txt = os.path.join(input_dir, filename)
            output_stl = os.path.join(output_dir, os.path.splitext(filename)[0] + ".stl")
            file_pairs.append((input_txt, output_stl))

    # Use multiprocessing to process files in parallel
    num_workers = min(cpu_count(), len(file_pairs))  # Limit workers to number of files
    with Pool(num_workers) as pool:
        pool.map(process_file, file_pairs)
```

## Предобработка точек валидации

```python
import os
import multiprocessing as mp
import pyvista as pv
import numpy as np
import fast_simplification

def simplify_and_save(input_stl_path, output_mesh_path, output_points_path, target_reduction=0.8):
    """
    Simplify a mesh from an STL file and save the simplified mesh and points.
    
    Parameters:
    - input_stl_path (str): Path to the input STL file.
    - output_mesh_path (str): Path to save the simplified STL file.
    - output_points_path (str): Path to save the points (as a .npy file).
    - target_reduction (float): Target reduction ratio (e.g., 0.8 for 80% reduction).
    """
    try:
        # Load the STL file using PyVista
        mesh = pv.read(input_stl_path)
        
        # Simplify the mesh using decimation
        simplified_mesh = fast_simplification.simplify_mesh(mesh, target_reduction=target_reduction)
        
        # Extract points from the simplified mesh
        points = simplified_mesh.points
        
        # Save the simplified mesh as STL
        simplified_mesh.save(output_mesh_path)
        
        # Save the points as a .npy file
        np.save(output_points_path, points)
        
        print(f"Successfully processed: {os.path.basename(input_stl_path)}")
    except Exception as e:
        print(f"Error processing {os.path.basename(input_stl_path)}: {e}")

def process_file(args):
    """
    Wrapper function to process a single file.

    Parameters:
    - args (tuple): A tuple containing the arguments needed for processing.
    """
    input_stl_path, output_mesh_path, output_points_path, target_reduction = args
    simplify_and_save(input_stl_path, output_mesh_path, output_points_path, target_reduction)

def main(input_dir, output_dir, target_reduction=0.8):
    # Get list of .stl files in the input directory
    stl_files = [f for f in os.listdir(input_dir) if f.lower().endswith('.stl')]
    
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # Prepare arguments for each file
    tasks = []
    for file_name in stl_files:
        input_stl_path = os.path.join(input_dir, file_name)
        output_mesh_path = os.path.join(output_dir, file_name)  # Save with the same filename
        output_points_filename = os.path.splitext(file_name)[0] + '_points.npy'
        output_points_path = os.path.join(output_dir, output_points_filename)
        tasks.append((input_stl_path, output_mesh_path, output_points_path, target_reduction))
    
    # Use multiprocessing Pool to process files in parallel
    with mp.Pool(processes=mp.cpu_count()) as pool:
        pool.map(process_file, tasks)
```

```python
import os
import pandas as pd
import numpy as np
import glob
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import cv2
from PIL import Image

import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import cv2
from PIL import Image

class ImageRegressionDataset(Dataset):
    def __init__(self, 
                 image_dir , 
                 transform=None):
        self.image_dir = image_dir
        self.transform = transform
        
        self.pairs = []
        self.image_folders = [f for f in os.listdir(image_dir)]

    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, idx):
        img_folder = self.image_folders[idx]
        
        img_path = os.path.join(self.image_dir, img_folder)
        
        images = []
        for img in glob.glob(img_path + '/*.png'):
            if 'spherical' in img or 'cylinder' in img:
                images.append(Image.open(img).convert('RGB'))
            elif 'up' in img or 'left' in img or 'front' in img:
                images.append(Image.open(img).convert('L'))
            else: continue
        
        if images is None:
            raise ValueError(f"Error loading image: {img_path}")

        if self.transform:
            images = [self.transform(img) for img in images]


        return torch.cat(images, dim = 0)
```

```python
import torch
from torch import nn
from torchvision import models

class EfficientNetRegression(nn.Module):
    def __init__(self):
        super(EfficientNetRegression, self).__init__()
        self.efficient_net = models.efficientnet_b1(pretrained=True)

        # Modify the first convolutional layer to accept 6 channels
        original_conv = self.efficient_net.features[0][0]
        self.efficient_net.features[0][0] = nn.Conv2d(
            in_channels=9,  
            out_channels=original_conv.out_channels,
            kernel_size=original_conv.kernel_size,
            stride=original_conv.stride,
            padding=original_conv.padding,
            bias=original_conv.bias is not None
        )

        # Modify the classifier to output a single value for regression
        self.efficient_net.classifier[1] = nn.Linear(
            self.efficient_net.classifier[1].in_features, 1
        )

    def forward(self, x):
        return self.efficient_net(x)
```

```python
model = models.resnet18(pretrained=True)

# Modify the first convolutional layer to accept 6 channels
# Original: Conv2d(3, 64, kernel_size=7, stride=2, padding=3, bias=False)
# New: Conv2d(6, 64, kernel_size=7, stride=2, padding=3, bias=False)

# Get the original first conv layer
original_conv = model.conv1

# Create a new conv layer with 6 input channels
new_conv = nn.Conv2d(9, original_conv.out_channels, kernel_size=original_conv.kernel_size,
                     stride=original_conv.stride, padding=original_conv.padding, bias=original_conv.bias)

# Initialize the new conv layer's weights by copying the original weights and duplicating for the additional channels
with torch.no_grad():
    new_conv.weight[:, :3, :, :] = original_conv.weight
    new_conv.weight[:, 3:6, :, :] = original_conv.weight  # Duplicate weights for the additional channels
    new_conv.weight[:, 6:9, :, :] = original_conv.weight
# Replace the model's conv1 with the new conv layer
model.conv1 = new_conv

# Modify the fully connected layer for regression
model.fc = nn.Linear(model.fc.in_features, 1)

# Move the model to the appropriate device
```

```python
import torch

def remove_module_prefix(state_dict):
    new_state_dict = {}
    for key, value in state_dict.items():
        if key.startswith('module.'):
            new_key = key[len('module.'):]  # Remove 'module.' prefix
        else:
            new_key = key
        new_state_dict[new_key] = value
    return new_state_dict

# Path to the downloaded model file
model_path = f"{effnetb1full_pytorch_default_1_path}/modelefficientnetb10.884312_0.942337.pth"

# Load the checkpoint
checkpoint = torch.load(model_path)

# Remove the 'module.' prefix from checkpoint keys
cleaned_checkpoint = remove_module_prefix(checkpoint)
```
